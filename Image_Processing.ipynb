# **Converting RGB images to Grayscale images using OpenCv**

```python
import cv2
```

```python
img=cv2.imread('/content/images.jpeg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```
