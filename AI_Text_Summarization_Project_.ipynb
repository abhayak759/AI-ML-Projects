```python
from transformers import pipeline               #here we have instaled all the dependencies required
from bs4 import BeautifulSoup     # it is a library that allows us to performm web scraping easily
import requests                    # allows to make https call out to the web posts and bring out the results and pass that result to beautifilsop
```

```python
summarizer = pipeline('summarization')
```

```python
URL="https://www.health.harvard.edu/blog"
#URL="https://hackernoon.com/will-the-game-stop-with-gamestop-or-is-this-just-the-beginning-2j1x32aa"
```

```python
r=requests.get(URL)   #this is grabbing the all the data in webpage the metadata, and it is going to be inside this r
```

```python
r
```

```python
r.text
```

```python
soup=BeautifulSoup(r.text, 'html.parser')
results=soup.find_all(['h1', 'p'])
```

```python
results
```

```python
soup
```

```python
text=[result.text for result in results] #looping through each result in each result
ARTICLE= ' '.join(text)     #it is joing the results together
```

```python
ARTICLE
```

```python
ARTICLE= ARTICLE.replace('.', '.<eos>')
ARTICLE= ARTICLE.replace('!','!<eos>')
ARTICLE= ARTICLE.replace('?', '?<eos>');
sentences=ARTICLE.split('<eos>')
```

```python
sentences
```

```python
max_chunk=500
current_chunk=0
chunks=[]
for sentence in sentences:       #looping through each of the sentences
  if len(chunks)==current_chunk+1:
    if len(chunks[current_chunk])+len(sentence.split(' '))<=max_chunk:
      chunks[current_chunk].extend(sentence.split(' '))
    else:
      current_chunk+=1
      chunks.append(sentence.split(' '))
  else:
    print(current_chunk)
    chunks.append(sentence.split(' '))
```

```python
for chunk_id in range(len(chunks)):
  chunks[chunk_id]= ' '.join(chunks[chunk_id])
```

```python
len(chunks[0].split(' '))
```

```python
res = summarizer(chunks, max_length=120, min_length=30, do_sample=False )
```

```python
res[0]['summary_text']
```

```python
' '.join([summ['summary_text'] for summ in res])
```

```python
text=' '.join([summ['summary_text'] for summ in res])
with open('Summarized Text.txt', 'w') as f:                #creates a text file that stores the summarized part of the blog or the article
  f.write(text)
```
